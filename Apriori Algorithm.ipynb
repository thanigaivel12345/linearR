```python
basket=df.groupby(["Member_number","itemDescription"])["item_count"].sum().unstack().reset_index().fillna(0).set_index("Member_number")
basket
```

```python
def encode_unit(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
basket_set=basket.applymap(encode_unit)
basket_set.head()
```

```python
from mlxtend.frequent_patterns import apriori
frequent_set=apriori(basket_set,min_support=0.07,use_colnames=True)
```

```python
frequent_set
```

```python
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_set,metric="lift",min_threshold=0.5)
rules
```

```python
rules[(rules["lift"]>=1) & (rules["confidence"]>=0.5)].head()
```
