```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("titanic.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df=df.drop(["Name","Sex","Age","Ticket","Cabin","Embarked"],axis=1)
```

```python
df
```

```python
x=df.drop(["Survived"],axis=1)
y=df.Survived
```

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5)
kf.get_n_splits(x,y)
```

```python
from sklearn.metrics import accuracy_score
model=DecisionTreeClassifier()
for train_ind,test_ind in kf.split(x,y):
    model.fit(x.iloc[train_ind,:],y.iloc[train_ind])
    pred=model.predict(x.iloc[test_ind,:])
print(round(accuracy_score(y.iloc[test_ind],pred),3))
```

```python
cv=cross_val_score(DecisionTreeClassifier(),x,y,cv=5,scoring="accuracy")
```
