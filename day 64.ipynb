```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

df=pd.read_csv("winequality-white.csv",sep=";")
df.head()
```

```python
df.dtypes
```

```python
x = df.drop(columns=['alcohol'],axis=1)
```

```python
y=df.alcohol
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=1)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
data = {'Model Name': [], 'MAPE': [],'RMSE':[],'RMSLE':[]}
df1 = pd.DataFrame(data)
df1
```

```python
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
model_ler = LinearRegression().fit(x_train,y_train)
pred = model_ler.predict(x_test)
```

```python
import math
from scipy import stats

rmse = np.sqrt(np.mean((y_test - pred)**2))

mape = (np.mean(np.abs(((y_test - pred)/y_test))))*100

rmsle = np.sqrt(np.square(np.log(pred + 1) - np.log(y_test + 1)).mean())

df2 = {'Model Name': 'Linear regression', 'MAPE':mape,'RMSE':rmse,'RMSLE':rmsle}
df = df.append(df1, ignore_index = True)
df
```

```python
# svm 
from  sklearn.svm import SVR
model_svr = SVR().fit(x_train,y_train)
pred = model_svr.predict(x_test)

# RMSE
rmse = np.sqrt(np.mean((y_test - pred)**2))

# MAPE
mape = (np.mean(np.abs(((y_test - pred)/y_test))))*100

#RMSLE
rmsle = np.sqrt(np.square(np.log(pred + 1) - np.log(y_test + 1)).mean())

df2 = {'Model Name': 'SVM regression', 'MAPE':mape,'RMSE':rmse,'RMSLE':rmsle}
df = df.append(df1, ignore_index = True)
df
```

```python
#ridge regression
from sklearn.linear_model import Ridge
ridgeReg = Ridge(alpha=0.0005, normalize=True)
ridgeReg.fit(x_train,y_train)
pred = ridgeReg.predict(x_test)

# RMSE
rmse = np.sqrt(np.mean((y_test - pred)**2))

# MAPE
mape = (np.mean(np.abs(((y_test - pred)/y_test))))*100

#RMSLE
rmsle = np.sqrt(np.square(np.log(pred + 1) - np.log(y_test + 1)).mean())

df2 = {'Model Name': 'Ridge regression', 'MAPE':mape,'RMSE':rmse,'RMSLE':rmsle}
df = df.append(df1, ignore_index = True)
df
```

```python
# lasso regression
from sklearn.linear_model import Lasso
lassoReg = Lasso(alpha=0.0005, normalize=True)
lassoReg.fit(x_train,y_train)
pred = lassoReg.predict(x_test)

# RMSE
rmse = np.sqrt(np.mean((y_test - pred)**2))

# MAPE
mape = (np.mean(np.abs(((y_test - pred)/y_test))))*100

#RMSLE
rmsle = np.sqrt(np.square(np.log(pred + 1) - np.log(y_test + 1)).mean())

df2 = {'Model Name': 'Lasso regression', 'MAPE':mape,'RMSE':rmse,'RMSLE':rmsle}
df = df.append(df1, ignore_index = True)
df
```

```python
from sklearn.tree import DecisionTreeRegressor

dtreg= DecisionTreeRegressor(max_depth=5)
dtreg.fit(x_train,y_train)
pred = dtreg.predict(x_test)

# RMSE
rmse = np.sqrt(np.mean((y_test - pred)**2))

# MAPE
mape = (np.mean(np.abs(((y_test - pred)/y_test))))*100

#RMSLE
rmsle = np.sqrt(np.square(np.log(pred + 1) - np.log(y_test + 1)).mean())

df2= {'Model Name': 'Decision Tree regression', 'MAPE':mape,'RMSE':rmse,'RMSLE':rmsle}
df = df.append(df1, ignore_index = True)
df
```
